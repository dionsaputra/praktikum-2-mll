# Access and Preprocess Dataset

```python
import pandas as pd
import numpy as np
# Library for Plot
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import rcParams
# Library for Resampling
from sklearn.utils import resample
```

```python
# Path menuju file dataset training
file_path_train = 'train.csv'

# Header kolom
headers = ['Age', 'Sex', 'Cp', 'Trestbps','Chol', 'FBS', 'Restecg', 'Thalach', 'Exang', 'Oldpeak', 'Slope', 'Ca' ,'Thal', 'Num']

# Membaca dataset training
train_dataset = pd.read_csv(file_path_train, header=None, engine='python', skiprows=1, na_values='?', names=headers)
train_dataset.head(15)
```

```python
# Path menuju file dataset training
file_path_test = 'test.csv'

# Header kolom
headers = ['Age', 'Sex', 'Cp', 'Trestbps','Chol', 'FBS','Restecg', 'Thalach', 'Exang','Oldpeak','Slope', 'Ca' ,'Thal', 'Num']

# Membaca dataset training
test_dataset = pd.read_csv(file_path_test, header=None, engine='python', skiprows=1, na_values='?', names=headers)
test_dataset.head(15)
```

Karena dataset imbalance, maka perlu diperhatikan oversampling dan undersampling pada data train

```python
# separate minority and majority classes
normal = train_dataset[train_dataset.Num==0]
sick = train_dataset[train_dataset.Num==1]

# upsample minority
sick_upsampled = resample(sick,
                          replace=True, # sample with replacement
                          n_samples=len(normal), # match number in majority class
                          random_state=23) # reproducible results

# combine majority and upsampled minority
train_dataset_oversampled = pd.concat([normal, sick_upsampled])

# check new class counts
train_dataset_oversampled.Num.value_counts()
```

```python
# separate minority and majority classes
normal = train_dataset[train_dataset.Num==0]
sick = train_dataset[train_dataset.Num==1]

# downsample majority
normal_downsampled = resample(normal,
                              replace = False, # sample without replacement
                              n_samples = len(sick), # match minority n
                              random_state = 23) # reproducible results

# combine minority and downsampled majority
train_dataset_downsampled = pd.concat([normal_downsampled, sick])

# checking counts
train_dataset_downsampled.Num.value_counts()
```
